```python
import coremltools

out_labels = ['投籃(shoot)', '運球(dribble)', '上籃(layup)']
coreml_model = coremltools.converters.keras.convert('models/my_model_90.h5', 
                                                    input_names=['image'],  
                                                    class_labels=out_labels,
                                                    output_names=['BasketballBConfidence'],
                                                   image_input_names=['image'],
                                                    image_scale=1./255.)
coreml_model.author = 'Funskie'
coreml_model.license = 'nope'
coreml_model.short_description = 'Basketball movement Recognition'
coreml_model.input_description['image'] = 'Image of Basketball'
coreml_model.output_description['BasketballBConfidence'] = 'Confidence and label of predicted Basketball movement'
coreml_model.output_description['classLabel'] = 'Label of predicted movement'
# Saving the Core ML model to a file.
print(coreml_model)
```

```python
coreml_model.save('coreml_InceptionV3_3.mlmodel')
```
